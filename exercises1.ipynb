```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import statsmodels as sm
```

```python
# data for now
data = {"state": ["Ohio", "Ohio", "Ohio", "Nevada", "Nevada", "Nevada"],
        "year": [2000, 2001, 2002, 2001, 2002, 2003],
        "pop": [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}
display(pd.DataFrame(data))
```

```python
# indexing
frame = pd.DataFrame(data)
frame["debt"] = np.arange(6.)
frame.loc[frame["state"] == "Ohio", ["debt", "pop"]] = 0.0
frame
```

```python
# sorting and ranking
frame = pd.DataFrame(data)
frame["rank"] = frame["pop"].rank(method="first", ascending=False)
frame.sort_values("pop", ascending=False)
```

```python
# aggregates
frame = pd.DataFrame(data)
frame["sum"] = frame[["pop", "year"]].sum(axis="columns")
display(frame)
frame.describe()
```

```python
# uniques
frame = pd.DataFrame(data)
display(frame["state"].unique())
display(frame["year"].value_counts())
```
